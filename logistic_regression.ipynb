```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
sumo=pd.read_csv('winequality-red.csv')
sumo.head(n=10)
```

```python
col=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality']
sumo=pd.read_csv('winequality-red.csv',names=col,header=1)
sumo.head()
```

```python
sumo.tail()
```

```python
sumo.count()
```

```python
sumo.describe()
```

```python
feature=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']
```

```python
X = sumo[feature]
Y = sumo.quality
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=0)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train,Y_train)
```

```python
Y_pred=log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_pred)
```
