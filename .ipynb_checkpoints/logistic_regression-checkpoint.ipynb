```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sumo=pd.read_csv('winequality-red.csv')
sumo.head(n=10)
```

```python
col=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality']
sumo=pd.read_csv('winequality-red.csv',names=col,header=1)
sumo.head()
```

```python
sumo.tail()
```

```python
sumo.count()
```

```python
range=[]
for i in sumo['quality']:
    if i>=3 and i<=5 :
        range.append('1')
    elif i>5 and i<=8 :
        range.append('2')
sumo['Range']=range        
```

```python
sumo.head()
```

```python
sumo.describe()
```

```python
X=sumo.iloc[:,0:11]
y=sumo.iloc[:,-1]
```

```python
bestfeatures = SelectKBest(score_func=chi2,k=10)
fit = bestfeatures.fit(X,y)
dfscores=pd.DataFrame(fit.scores_)
dfcolumns=pd.DataFrame(X.columns)
featureScores=pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']
print(featureScores.nlargest(10,'Score'))
```

```python
feature=['fixed acidity','citric acid','chlorides','total sulfur dioxide','density','pH','sulphates','alcohol', 'quality']
```

```python
X = sumo[feature]
Y = sumo.Range
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=0)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train,Y_train)
```

```python
Y_pred=log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_pred)
```
